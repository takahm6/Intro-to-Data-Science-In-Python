```python
def convert_housing_data_to_quarters():
    '''Converts the housing data to quarters and returns it as mean 
    values in a dataframe. This dataframe should be a dataframe with
    columns for 2000q1 through 2016q3, and should have a multi-index
    in the shape of ["State","RegionName"].
    
    
    Note: Quarters are defined in the assignment description, they are
    not arbitrary three month periods.
    
    The resulting dataframe should have 67 columns, and 10,730 rows.
    '''
    # Load the excel file
    housingDF = pd.read_csv('City_Zhvi_AllHomes.csv')
    #stateDF =  pd.DataFrame(states.items(), columns=['State', 'State Name'])
    stateDF = pd.DataFrame.from_dict(list(states.items()))
    stateDF.rename(columns={0:'State', 1:'State Name'}, inplace=True)
                                     
    
    # Merge it with the State name
    housePrice = pd.merge(stateDF, housingDF, on='State')
    
    # drop columns that are not needed
    housePrice.drop(housePrice.columns[4:52], axis=1, inplace=True)
    housePrice.drop(housePrice.columns[205:], axis=1, inplace=True)
    cols = ['State', 'RegionID']
    housePrice.drop(cols, axis=1, inplace=True)
    
    # change column names to what had been specified
    housePrice.rename(columns={'State Name':'State'}, inplace=True)
    
    # create multiindex
    housePrice = housePrice.set_index(['State', 'RegionName'])
    
    # convert the column labels to datetime
    from datetime import datetime
    from dateutil.parser import parse
    
    cols = housePrice.columns.tolist()
    housePrice.columns = map(lambda x: datetime.strptime(x, '%Y-%m'), cols)
    
    # Resample by quota and calculate the mean
    housing=housePrice.resample('Q', axis=1).mean()
    
    
    # Rename the columns back to string
    housing = housing.rename(columns=lambda x: str(x.to_period('Q')).lower())
    
    return housing

convert_housing_data_to_quarters()
```

```python
def run_ttest():
    '''First creates new data showing the decline or growth of housing prices
    between the recession start and the recession bottom. Then runs a ttest
    comparing the university town values to the non-university towns values, 
    return whether the alternative hypothesis (that the two groups are the same)
    is true or not as well as the p-value of the confidence. 
    
    Return the tuple (different, p, better) where different=True if the t-test is
    True at a p<0.01 (we reject the null hypothesis), or different=False if 
    otherwise (we cannot reject the null hypothesis). The variable p should
    be equal to the exact p value returned from scipy.stats.ttest_ind(). The
    value for better should be either "university town" or "non-university town"
    depending on which has a lower mean price ratio (which is equivilent to a
    reduced market loss).'''
    
    from scipy import stats
    
    hdf = convert_housing_data_to_quarters()
    rec_start=get_recession_start()
    rec_bottom = get_recession_bottom()
    ul = get_list_of_university_towns()
    
    # adding a new column to show the house price ratio between before recession and the recession bottom
    hdf['PriceRatio'] = hdf.iloc[:, hdf.columns.get_loc(str(rec_start).lower())-1].div(hdf[str(rec_bottom).lower()])
    
    # dividing the set ino two groups
    # convert to a list of tuples
    univTown_list = ul.to_records(index=False).tolist()
    
    univGroup = hdf.loc[univTown_list]
    
    otherGroup = hdf.loc[~hdf.index.isin(univTown_list)]
    
    # Run a ttest
    test_result= stats.ttest_ind(univGroup['PriceRatio'], otherGroup['PriceRatio'], nan_policy='omit')
    different = (test_result.pvalue < 0.01)
    p = test_result.pvalue
    if univGroup['PriceRatio'].mean() > otherGroup['PriceRatio'].mean(): 
        better = 'non-university town'
    else:
        better = 'university town'


    
    return (different, p, better)

run_ttest()
```
